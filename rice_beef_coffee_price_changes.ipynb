```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/SANTOSTAVARES/cerial-prices-changes-within-last-30-years/e2e6c81777500772decb01e5a3a8aa6c3fc874e6/rice_wheat_corn_prices.csv'
df = pd.read_csv(url)
```

```python
df["Year"] = df["Year"].astype(str)
```

```python
semester_list = []
first_half = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

for i in df.itertuples():
  if i.Month in first_half:
    semester_list.append(f'{i.Year}-1')
  else:
    semester_list.append(f'{i.Year}-2')
df['Semester'] = semester_list
```

```python
wheat_df = df[["Semester", "Year", "Month", "Price_wheat_ton", "Inflation_rate", "Price_wheat_ton_infl"]]

rice_df = df[["Year", "Month", "Price_rice_ton", "Inflation_rate", "Price_rice_ton_infl"]]

corn_df = df[["Year", "Month", "Price_corn_ton", "Inflation_rate", "Price_corn_ton_infl"]]
```

```python
everage = []
period = df['Semester'].unique()
for x in period:
  price_sum = []
  for i in df.itertuples():
    if i.Semester == x:
      price_sum.append(i.Price_wheat_ton)
  everage.append(sum(price_sum)/len(price_sum))
```
